```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
x_original = torch.tensor([[1.0, 5.0, 10.0]])


model = nn.Sequential(
    # encoder
    nn.Linear(3, 2),
    nn.ReLU(),

    # decoder
    nn.Linear(2, 3)
)


criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)


# training
for epoch in range(1001):
    output = model(x_original) # output from decoder
    loss = criterion(output, x_original)

    optimizer.zero_grad() # in pytorch, the weight is not removed automatically, so reset it every epoch.
    loss.backward() # backpropagation
    optimizer.step() # update weight

    if epoch % 200 == 0:
        print(f"Epoch {epoch} - (Loss): {loss.item():.6f}")

# 4. 최종 결과 확인
final_output = model(x_original).detach()

print(f"original : {x_original.numpy()[0]}")
print(f"output: {final_output.numpy()[0]}")
```
