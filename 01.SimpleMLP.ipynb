```python
import torch
import torch.nn as nn
import numpy as np
```

## two layer nn

```python
# x -> h1 -> h2 -> y
x = 1.0

w1 = 0.5    # h1 weight
w2 = 0.8    # h2 weight

y_true = 1.0 # true value

learning_rate = 0.1

# training loop : forward and Backward !!
# we try loop 1 : if loop more and more then loss values is decreasing
for epoch in range(3):

    # --- [Forward Pass] ---
    h = x * w1
    y_hat = h * w2

    loss = 0.5 * (y_hat - y_true)**2 # 0.5 = 1/2 : depends on number of x, 1/2n. 2 is for differential

    print(f"\n--- Forward result ---")
    print(f"y_hat: {y_hat:.2f}, Loss: {loss:.4f}")



    # --- [Backward] ---
    d_loss_y_hat = y_hat - y_true # effect of y_hat for Loss

    # dL/dw2 = dL/dy_hat * dy_hat/dw2 = dL/dy_hat * h
    d_loss_w2 = d_loss_y_hat * h # (-0.6 * 0.5 = -0.3)

    # dL/dw1 = dL/dy_hat * dy_hat/dh * dh/dw1
    # dy_hat/dh = w2,  dh/dw1 = x
    d_loss_w1 = d_loss_y_hat * w2 * x # (-0.6 * 0.8 * 1.0 = -0.48)

    print(f"\n--- Backward 결과 (Gradients) ---")
    print(f"w2 Gradient: {d_loss_w2:.2f}")
    print(f"w1Gradient: {d_loss_w1:.2f}")



    # --- [Weight Update] ---
    w2 = w2 - (learning_rate * d_loss_w2)
    w1 = w1 - (learning_rate * d_loss_w1)

    print(f"\n--- new weights ---")
    print(f"new w1: {w1:.4f}, new w2: {w2:.4f}")
```

```python
class SimpleMLP(nn.Module): #in nn module has function to calculate backpropagation
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(3, 2) # hidden layer 1 makes data 3->2
        self.fc2 = nn.Linear(2, 1) # hidden layer 2 makes data 2->1

        self.relu = nn.ReLU()

    def forward(self, x):

        x = self.fc1(x)
        x = self.relu(x)

        x = self.fc2(x)
        return x
```

```python
model = SimpleMLP()


sample_input = torch.tensor([[1.0, 2.0, 3.0]])

output = model(sample_input)

print(f"input shape: {sample_input.shape}") # torch.Size([1, 3])
print(f"output shape: {output.shape}")       # torch.Size([1, 1])
```
